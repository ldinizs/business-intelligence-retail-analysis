```python
import pandas as pd

# Carregar os dados
# Se o arquivo estiver na mesma pasta que o notebook:
df = pd.read_excel('Data Source.xlsx')

print("--- ANÁLISE DE QUALIDADE INICIAL ---")

# 1. Verificar Valores Nulos
null_count = df.isnull().sum()
print(f"\nValores Nulos por Coluna:\n{null_count[null_count > 0]}")
if null_count.sum() == 0:
    print("Nenhum valor nulo encontrado.")

# 2. Verificar Linhas Duplicadas
duplicate_count = df.duplicated().sum()
print(f"\nTotal de linhas duplicadas: {duplicate_count}")

# 3. Validar Formato de Datas (Tentativa de conversão)
try:
    test_dates_order = pd.to_datetime(df['Order Date'], errors='coerce')
    test_dates_ship = pd.to_datetime(df['Ship Date'], errors='coerce')
    
    invalid_order_dates = test_dates_order.isna().sum()
    invalid_ship_dates = test_dates_ship.isna().sum()
    
    print(f"\nDatas de Pedido Inválidas: {invalid_order_dates}")
    print(f"Datas de Envio Inválidas: {invalid_ship_dates}")
except Exception as e:
    print(f"Erro ao validar datas: {e}")
```

```python
# 1. Remover duplicatas e nulos
df_cleaned = df.drop_duplicates().dropna()

# 2. Corrigir inconsistências de data (Manter apenas datas válidas)
df_cleaned = df_cleaned[df_cleaned['Ship Date'] >= df_cleaned['Order Date']]

# 3. Feature Engineering: Criar coluna de tempo de despacho (Valor agregado para o BI)
df_cleaned['Days to Ship'] = (df_cleaned['Ship Date'] - df_cleaned['Order Date']).dt.days

# 4. Exportação Final para Excel (Para subir no Tableau)
df_cleaned.to_excel('Retail_Data_Cleaned.xlsx', index=False)

print("✅ Processo concluído!")
print(f"Registros finais salvos: {len(df_cleaned)}")
```
